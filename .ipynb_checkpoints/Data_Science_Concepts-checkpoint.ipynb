## Overfitting

The main objectives of this section are to:
- Show how an underfit model look like
- Show how an overfit model look like

This function will be used to illustrate the concept of overfitting and underfitting:
 $$y = 5x^3 + 2.5x^2 + 11 + \epsilon $$

```python
# Set the random seed for reproduceable output.
np.random.seed(123)

# Create synthetic data to to use for illustration

# Creates a range from -2 to 2, with 100 steps.
x = np.linspace(-1, 1, 100)*2

# Creates the target variable
y = 5*x**3 + 2.5*x**2 + 11 + np.random.randn(100)*5.0 
```

```python
# This the function above
def f(t):
    return 5*t**3 + 2.5*t**2 + 11

# Plot for the range of x we created above.
plt.figure()
plt.scatter(x, y, c="cornflowerblue", label = "data") # Plot the points
plt.plot(x, f(x), 'b--', label="function") # Plot the driving function (no error)
plt.xlabel("x (independent variable)") 
plt.ylabel("y (target)")
plt.title("Plotted Synthetic Data")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Reshape our x and y
x_reshaped = x.reshape(-1,1)
y_reshaped = y.reshape(-1,1)

# Instantiate and fit the linear regression model
lr = LinearRegression()
lr_model = lr.fit(x_reshaped, y_reshaped)

# Score the model
lr.score(x_reshaped, y_reshaped)
```

```python
# Grab the prediction data
y_pred = lr_model.predict(x_reshaped)

# Plot
# Data and fit curve
plt.figure()
plt.scatter(x, y, label='True labels', c="cornflowerblue") # Plot the data
plt.plot(x, f(x), 'b--', label="Function", alpha=0.5) # Plot the driving function (no error)
plt.plot(x, y_pred, linewidth=2, color='r', linestyle='--', label='Model') # Plot the predictions
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Curve fitting')
# Show
plt.show()
```

Save the underfit model:

```python
from joblib import dump, load
dump(lr, 'saved_models/overfitting_underfit_model.joblib') 
```

```python
# The Tensorflow libraries we'll need.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.metrics import MeanAbsoluteError

# Feed forward NN.
model = Sequential()

# 40/20 alternating layers
model.add(Dense(40, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(40, activation='relu'))

# Single node output as this is regression
model.add(Dense(1)) # Default activation is linear, no need to define any activation function here

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(), 
             loss=tf.keras.losses.MeanAbsoluteError(), # MAE loss function
             metrics=[MeanAbsoluteError()]) # MAE evaluation metric
```

```python
# Train for a resonable number of epochs
history = model.fit(x, y, epochs=100, verbose=0)

# Append loss to running tally
# mae_list = mae_list + history.history['mean_absolute_error']

# Make predictions
y_pred = model.predict(x)

# Plot
# plt.subplots(1,2, figsize=(14, 4))

# Data and fit curve
# plt.subplot(1,2,1)
plt.scatter(x, y, label='True labels', c="cornflowerblue") # Plot the data
plt.plot(x, f(x), 'b--', label="Function", alpha=0.5) # Plot the driving function (no error)
plt.plot(x, y_pred, linewidth=2, color='r', linestyle='--', label='Model') # Plot the predictions
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Curve fitting')

# # Error for each epoch
# plt.subplot(1,2,2)
# plt.plot(history.history['mean_absolute_error'], c='cornflowerblue')
# plt.title('Model error')
# plt.xlabel('epoch')
# plt.ylabel('Mean Absolute Error (MAE)')

# Show
plt.show()
```
